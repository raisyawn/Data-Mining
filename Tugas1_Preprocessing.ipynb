#**Import Data**

```python
import pandas as pd
# Memuat data dari file CSV
df = pd.read_csv('/movie_sample_dataset.csv')
df.head()
```

#**Memeriksa Data**

```python
df.info()
```

```python
print(df.isnull().sum())
```

#**Membersihkan Data**

```python
df.loc[:, 'color'] = df['color'].str.strip().str.capitalize()
df.loc[:, 'country'] = df['country'].str.strip().str.upper()
df = df[(df['duration']>=0)&
              (df['gross']>=0)&
              (df['title_year']>=0)&
              (df['budget']>=0)&
              (df['imdb_score']>=0)&
              (df['movie_facebook_likes']>=0)]
df = df.replace('N/A', pd.NA)
df.dropna(inplace=True)
df
```

```python
print(df.isnull().sum())
```

```python
df.dtypes
```

```python
df['gross'] = df['gross'].astype('int')
df['budget'] = df['budget'].astype('int')
df.dtypes
```

```python
df['actors'] = df['actors'].str.replace(',', ', ', regex=False)
df['genres'] = df['genres'].str.replace('|', ', ', regex=False)
df
```

#**Menyimpan Data**

```python
df.to_csv('movie_sample_dataset_done.csv', index=False)
```
